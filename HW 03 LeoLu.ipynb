```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import numpy as np

# Load penguins dataset
url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv'
penguins = pd.read_csv(url)

# Helper function to calculate stats
def calculate_stats(df, column):
    mean = df[column].mean()
    median = df[column].median()
    std = df[column].std()
    iqr = df[column].quantile(0.75) - df[column].quantile(0.25)
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    min_val = df[column].min()
    max_val = df[column].max()
    return mean, median, std, iqr, q1, q3, min_val, max_val
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import numpy as np

# Load penguins dataset
url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv'
penguins = pd.read_csv(url)

# Drop rows with missing values in 'flipper_length_mm' and 'species'
penguins = penguins.dropna(subset=['flipper_length_mm', 'species'])

# Function to calculate statistics
def calc_stats(df):
    mean_val = df.mean()
    median_val = df.median()
    min_val = df.min()
    max_val = df.max()
    std_val = df.std()
    q1, q3 = df.quantile([0.25, 0.75])
    return mean_val, median_val, min_val, max_val, std_val, q1, q3

# Create subplots for histograms and KDE for each species
species_list = penguins['species'].unique()
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 6))

for i, species in enumerate(species_list):
    ax = axes[i]
    species_data = penguins[penguins['species'] == species]['flipper_length_mm']

    # Calculate stats
    mean_val, median_val, min_val, max_val, std_val, q1, q3 = calc_stats(species_data)
    
    # Create histogram using Plotly
    fig_histogram = go.Figure(data=[go.Histogram(x=species_data, nbinsx=20)])
    fig_histogram.update_layout(
        title=f"Flipper Length for {species}",
        xaxis_title="Flipper Length (mm)",
        yaxis_title="Count"
    )
    
    # Seaborn KDE
    sns.kdeplot(species_data, ax=ax, fill=True, color='skyblue', alpha=0.5)

    # Mark Mean, Median
    ax.axvline(mean_val, color='red', linestyle='--', label=f"Mean: {mean_val:.2f}")
    ax.axvline(median_val, color='green', linestyle='-', label=f"Median: {median_val:.2f}")

    # Mark IQR
    ax.axvspan(q1, q3, alpha=0.2, color='purple', label=f"IQR: [{q1:.2f}, {q3:.2f}]")

    # Mark Range
    ax.axvspan(min_val, max_val, alpha=0.1, color='orange', label=f"Range: [{min_val:.2f}, {max_val:.2f}]")

    # Mark Two Standard Deviations
    ax.axvspan(mean_val - 2*std_val, mean_val + 2*std_val, alpha=0.1, color='blue', label=f"2 Std Dev: [{mean_val - 2*std_val:.2f}, {mean_val + 2*std_val:.2f}]")
    
    ax.set_title(f'{species} KDE with Stats')
    ax.legend()

plt.tight_layout()
plt.savefig('penguins_flipper_length.png', dpi=300) 
```

# ChatGPT's response
Your observations are spot-on. Each of the three methods—box plots, histograms, and KDE plots—offers distinct advantages and disadvantages in visualizing data distribution. Here's a more formal breakdown in a "pros and cons" list:

### 1. **Box Plot**
**Pros**:
- **Summary Statistics**: Clearly shows the median, interquartile range (IQR), and outliers.
- **Outlier Detection**: Outliers are visualized effectively, making it great for spotting anomalies.
- **Compact Visualization**: Takes up little space, making it good for comparing multiple datasets side by side.
- **Non-Parametric**: Doesn’t assume any distribution shape (e.g., normality), just shows the spread and key statistics.

**Cons**:
- **No Frequency Information**: Does not show the frequency or shape of the distribution.
- **No Continuous Curve**: Doesn’t give a sense of the "smoothness" of the data distribution.
- **Limited Insight for Symmetry**: It may not give a complete picture of the skewness or symmetry of the data.

---

### 2. **Histogram**
**Pros**:
- **Frequency Representation**: Directly shows how often data points occur in each bin, making it intuitive to understand data distribution.
- **Distribution Shape**: Gives an idea of the distribution shape (e.g., normal, skewed, bimodal, etc.).
- **Bin Control**: Adjustable bin size can provide different levels of detail about the distribution.

**Cons**:
- **Discrete Representation**: The choice of bins can heavily influence the appearance, which may lead to misinterpretation (too few or too many bins).
- **No Summary Statistics**: Does not directly show median, quartiles, or outliers.
- **Not Continuous**: Data is represented in intervals, which can lead to loss of precision for continuous data.

---

### 3. **Kernel Density Estimator (KDE)**
**Pros**:
- **Smooth, Continuous Curve**: Provides a smooth estimate of the distribution, ideal for visualizing continuous data.
- **Visual Clarity**: Easier to see the overall shape of the distribution compared to histograms, especially with small datasets.
- **No Binning**: Unlike histograms, KDE doesn’t depend on arbitrary bin choices, reducing potential biases from bin size.

**Cons**:
- **Sensitive to Bandwidth**: The smoothness of the curve depends on the bandwidth (smoothing parameter), and poor choices can over- or under-smooth the data.
- **No Summary Statistics or Outliers**: Does not show median, quartiles, or outliers clearly like a box plot does.
- **Less Intuitive for Beginners**: Might be harder to interpret for those unfamiliar with the method.

---

### **Summary**
- **Box plots** are great for summarizing key statistics and identifying outliers but don’t show the distribution shape or frequency.
- **Histograms** provide an intuitive sense of frequency and distribution shape but can be sensitive to bin size and don’t represent outliers well.
- **KDE plots** offer a smooth, continuous visualization of the data distribution but lack direct representation of summary statistics or outliers.

Generate 1000 random samples with a=2 and scale=2 from the gamma distribution. The gamma distribution is usually right skewed

Create a histogram of the above gamma distribution

### sample2 = -stats.gamma(a=2,scale=2).rvs(size=1000)
Generate a set of inverse gamma-distributed samples, where positive gamma-distributed samples are right-skewed and the inverse are left-skewed

```python
from scipy import stats
import pandas as pd
import numpy as np
  
sample1 = stats.gamma(a=2, scale=2).rvs(size=1000)

fig1 = px.histogram(pd.DataFrame({'data': sample1}), x="data", title="Sample 1: Gamma Distribution")

mean_value1 = sample1.mean()
median_value1 = np.quantile(sample1, [0.5])

fig1.add_vline(x=mean_value1, line_dash="dash", line_color="red", 
                annotation_text="Mean", annotation_position="top right")
fig1.add_vline(x=median_value1, line_dash="dash", line_color="blue", 
                annotation_text="Median", annotation_position="top right")

fig1.show(renderer="png")
```

```python
sample2 = - stats.gamma(a=2,scale=2).rvs(size=1000)
fig2 = px.histogram(pd.DataFrame({'data': sample2}), x="data", title="Sample 2: Negated Gamma Distribution")

mean_value2 = sample2.mean()
median_value2 = np.quantile(sample2, [0.5])

fig2.add_vline(x=mean_value2, line_dash="dash", line_color="red", 
                annotation_text="Mean", annotation_position="top right")
fig2.add_vline(x=median_value2, line_dash="dash", line_color="blue", 
                annotation_text="Median", annotation_position="top right")

fig2.show(renderer="png")
```
